```python
# import libraries
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
flower=sns.load_dataset("iris")
flower
#draw a line plot
sns.lineplot(x="sepal_length", y="sepal_width", data=flower)
plt.title("morphological characteristic of flower")
plt.xlim(2)
plt.ylim(1)
plt.show()
```

```python
# import libraries
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Load dataset
flower=sns.load_dataset("iris")
flower
#draw a line plot
sns.lineplot(x="sepal_length", y="sepal_width", data=flower)
plt.title("morphological characteristic of flower")
# Set style
sns.set_style("dark")
sns.set_style(style=None, rc=None)

plt.show()
```

```python
## Size of figure
```

```python
# import libraries
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Load dataset
flower=sns.load_dataset("iris")
# change figure
plt.figure(figsize=(3,2))
#draw a line plot
sns.lineplot(x="sepal_length", y="sepal_width", data=flower)
plt.title("morphological characteristic of flower")

plt.show()
```
